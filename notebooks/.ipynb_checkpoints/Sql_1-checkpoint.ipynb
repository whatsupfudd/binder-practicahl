```haskell
:set -XOverloadedStrings

import Text.Blaze.Html5 as H
import Text.Blaze.Html5.Attributes as A
import Data.ByteString

dbPasswd = "fleurie" :: ByteString
```

```haskell
:set -XQuasiQuotes

import qualified Hasql.Connection as DbC
import qualified Hasql.Session as Se
import Hasql.TH as Th

import GHC.Word (Word16)
import qualified Data.Time.Clock as Clk
import qualified Data.Vector as Vc


let dbSettings = DbC.settings "ledna" 5432 "larose" dbPasswd "bblossom"

dbConn <- DbC.acquire dbSettings
```

```haskell
import Data.Int as DI
import Data.Text as DT

-- id, localegroup, label, description, locale
type FindCourseOut = (DI.Int32, DI.Int32, DT.Text, Maybe DT.Text, DT.Text)

findToMCourse :: Se.Session (Vc.Vector FindCourseOut)
findToMCourse =
  Se.statement () [Th.vectorStatement|
    select id::int4, localegroup::int4, label::text, description::text?, locale::text
    from crs.course_b
    order by localegroup desc
  |]


fcoHeader =
  th "ID" >> th "LclGrp" >> (th ! A.style "text-align: center;" $ "Label") >> (th ! A.style "text-align: center;" $ "Description") >> th "Locale"

fcoToHtml (id, localeGroup, label, mbDescription, locale) = do
  td (toHtml . show $ id)
  td (toHtml . show $ localeGroup)
  td ! A.style "text-align: left;" $ (toHtml label)
  case mbDescription of
    Nothing -> td ! A.style "text-align: center;" $ "---"
    Just aText -> td ! A.style "text-align: left;" $ (toHtml aText)
  td (toHtml locale)
```

```haskell
dbVal <- case dbConn of
  Right conn -> do
    rezA <- Se.run findToMCourse conn
    case rezA of
      Left err -> pure . Left $ show err
      Right mbVal -> pure $ Right mbVal
  Left err -> pure . Left $ show err

docTypeHtml $
    case dbVal of
      Left errMsg -> p (toHtml errMsg)
      Right vcVal -> do
         p "Courses:"
         table $ do
           fcoHeader
           Vc.forM_ vcVal (tr . fcoToHtml)
```
